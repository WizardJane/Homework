```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_x = np.linspace(1, 10, 20)
r = np.random.rand(1, 20)
data_y = data_x + r
```

```python
n = data_x.shape[0]
#y = b1 * x + b0

mean_x = data_x.mean()
mean_y = data_y.mean()
mean_xy = (data_x * data_y).mean()
mean_y2 = (data_y * data_y).mean()
mean_x2 = (data_x * data_x).mean()


b1 = (mean_xy - mean_x * mean_y) / (mean_x2 - mean_x * mean_x)
b0 = mean_y - b1 * mean_x
```

```python
lsm_x = np.arange(0, np.max(data_x) + 2, 1)
lsm_y = np.array(b1 * lsm_x + b0)

```

```python
plt.title("LSM")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(data_x, data_y, color="blue", label="data")
plt.plot(lsm_x, lsm_y, "--", c="grey", label="LSM")
plt.legend()
plt.show()
```
