```python
class Node:
    def __init__(self, x):
        self.val = x
        self.next = None
        self.prev = None
        self.id = None
```

```python
class Doubly_linked_list:
    def __init__(self):
        first = Node(None)
        last = Node(None)
        self.list = [first, last]
        self.head = self.list[0]
        self.end = self.list[1]
        self.head.next = 1
        self.head.id = 0
        self.end.prev = 0
        self.end.id = 1
        
    

    def AddToEnd(self, node):
        prev_node = self.list[self.end.prev]
        #привязываем новый элемент к старым
        node.next = 1 
        node.prev = self.end.prev
        #добавляем новый элемент в массив
        id = len(self.list)
        node.id = id
        self.list.append(node)
        #привязываем старые элементы к новому
        prev_node.next = id
        self.end.prev = id

    def AddToBeginning(self, node):
        next_node = self.list[self.head.next]
        #привязываем новый элемент к старым
        node.next = self.head.next
        node.prev = 0
        #добавляем новый элемент в массив
        id = len(self.list)
        node.id = id
        self.list.append(node)
        #привязываем старые элементы к новому
        next_node.prev = id
        self.head.next = id

    def insert(self, newnode, id):
        if id == 0:
            self.AddToBeginning(newnode)
            return
        if id >= len(self):
            self.AddToEnd(newnode)
            return
        
        node = self.list[self.head.next]
        i = 0
        while i < id:
            node = self.list[node.next]
            i += 1
        
        newnode.id = len(self.list)
        newnode.prev = node.prev
        newnode.next = node.id
        node.prev = newnode.id
        self.list.append(newnode)
        self.list[newnode.prev].next = newnode.id


    def get(self, newnode):
        i = 0
        node = self.list[self.head.next]
        while (node.val != newnode.val and node.id != 1):
            node = self.list[node.next]
            i += 1
        if (node.val == newnode.val):
            return i

    def contains (self, newnode):
        node = self.list[self.head.next]
        while (node.val != newnode.val and node.id != 1):
            node = self.list[node.next]
        if (node.val == newnode.val):
            return True
        
        return False
    
    def remove(self, newnode):
        node = self.list[self.head.next]
        while (node.val != newnode.val and node.id != 1):
            node = self.list[node.next]
        if (node.val == newnode.val):
            (self.list[node.next]).prev = node.prev
            (self.list[node.prev]).next = node.next

    def __len__(self):
        n = 0
        node = self.list[self.head.next]
        while (node != self.end):
            n += 1
            node = self.list[node.next]
        return n

    def __str__(self):
        s = "["
        node = self.list[self.head.next]
        while (node != self.end):
            s += str(node.val) + ", "
            node = self.list[node.next]
            
        
        if (len(s) > 2):
            s = s[:-2]

        return s + "]"
```
