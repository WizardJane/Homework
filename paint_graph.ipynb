```python
#Ввод

n = int(input()) #количество вершин
g = [[] for i in range(n)] #хранение рёбер графа
print("n =", n)
print("edges:")
x = int(input()) #ввод количества рёбер
for i in range(x):
    a, b = map(int, input().split())
    g[a].append(b)
    g[b].append(a)
    print("(" + str(a) + "," + str(b) + ")")


m = int(input()) #количество цветов
print("m =", m)

color = [0] * n #массив для хранения цветов
q = [] #массив для хранения пути вершин

#v - номер текущей вершины
def dfs(v): 
    global g, m, n, color, q

    if color[v] != 0:
        return True
    q.append(v)
    #перебираем цвета для вершины
    for c in range(1, m + 1):    
        #проверяем, чтобы у соседей не было такого цвета
        a = True
        for u in g[v]:           
            if color[u] == c:
                a = False
                break
        if not a:
            continue

        
        color[v] = c
        p = True
        for u in g[v]:
            p *= dfs(u)
        if p:
            return True
        color[v] = 0 #в случае, когда поддерево не получилось раскрасить, зачищаем его
        i = len(q) - 1
        while q[i] != v:
            color[q[i]] = 0
            q.pop()
            i -= 1

    return False



def find_answer():
    global color, n
    ans = True
    for i in range(n):
        if color[i] == 0:
            ans *= dfs(i)

    return ans

print("answer:", end=' ')
if find_answer():
    print("true")
else:
    print("false")

#print(color)

```
